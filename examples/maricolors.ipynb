# Mariprism Colors


### Here we show the colors and colormaps available on Mariprism!

```python
# Import general
import sys
import numpy as np
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


# Import local
sys.path.append( '../mariprism' )
from _marigradients import plot_colortable, make_gradient
from _maricolors_dict import FLOWER_BLOSSOM, FLOWER_RAD, ICY_FLOWER
from _maricolors_dict import GAS_FILAMENTS, FOREST_TROPICAL, SUCULENT_GARDEN 
from _maricolors_dict import mari_BLUES, mari_PURPLES
```

```python
# Generate data for test plot
data = np.outer(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))
data = (data / data.max() - 0.5) * 5.0
```

### 1. Examples of colors

```python
# Example 1.
name = 'FLOWER BLOSSOM'
plot_colortable(FLOWER_BLOSSOM)
plt.show()

colormarip = make_gradient(FLOWER_BLOSSOM, name='flower_blossom')
plt.figure(figsize=(4, 3))
plt.pcolormesh(data, cmap=colormarip)
plt.colorbar(aspect=5)
plt.title(str(name))
plt.show()

# Example 2.
name = 'FLOWER RAD'
plot_colortable(FLOWER_RAD)
plt.show()

colormarip = make_gradient(FLOWER_RAD, name='flower_rad')
plt.figure(figsize=(4, 3))
plt.pcolormesh(data, cmap=colormarip)
plt.colorbar(aspect=5)
plt.title(str(name))
plt.show()

# Example 3.
name = 'ICY FLOWER'
plot_colortable(ICY_FLOWER)
plt.show()

colormarip = make_gradient(ICY_FLOWER, name='icy_flower')
plt.figure(figsize=(4, 3))
plt.pcolormesh(data, cmap=colormarip)
plt.colorbar(aspect=5)
plt.title(str(name))
plt.show()
```
