# Here we present the colors available on Mariprism!

```python
import sys
sys.path.append( '../mariprism' )
import numpy as np
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from _maricolors import FLOWER_BLOSSOM, FLOWER_RAD, ICY_FLOWER #available color combos
from _marigradients import plot_colortable, make_gradient

data = np.random.rand(100,100)
```

```python
plot_colortable(FLOWER_BLOSSOM)
plt.show()

flower_blossom = make_gradient(FLOWER_BLOSSOM, name='flower_blossom')
plt.figure(figsize=(4, 3))
plt.pcolormesh(data, cmap=flower_blossom)
plt.colorbar()
plt.show()

make_gradient(FLOWER_BLOSSOM, name='flower_blossom')
```

```python
plot_colortable(FLOWER_RAD)
plt.show()

flower_rad = make_gradient(FLOWER_RAD, name='flower_rad')
plt.figure(figsize=(4, 3))
plt.pcolormesh(data, cmap=flower_rad)
plt.colorbar()
plt.show()

```
