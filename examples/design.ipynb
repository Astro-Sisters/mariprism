# mariprism

```python
import sys
sys.path.append( '../mariprism' )

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
from _maricolors import FLOWER_RAD, FLOWER_BLOSSOM, ICY_FLOWER
from _marigradients import make_gradient
```

```python
data = np.random.rand(100,100)
```

```python
# make discret colors from a colormap
new_inferno = cm.get_cmap('inferno', 5)# visualize with the new_inferno colormaps
plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = new_inferno)
plt.colorbar()
plt.show()

new_inferno = cm.get_cmap('inferno', 2)# visualize with the new_inferno colormaps
plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = new_inferno)
plt.colorbar()
plt.show()
```

```python
# create colormap in subrange of existing one
# modified hsv in 256 color class
hsv_modified = cm.get_cmap('hsv', 256)# create new hsv colormaps in range of 0.3 (green) to 0.7 (blue)
newcmp = ListedColormap(hsv_modified(np.linspace(0.3, 0.7, 256)))# show figure
plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = newcmp)
plt.colorbar()
plt.show()

newcmp = ListedColormap(hsv_modified(np.linspace(0.0, 1.0, 256)))# show figure
plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = newcmp)
plt.colorbar()
plt.show()
```

```python
# create colormap combining two sequentials

# define top and bottom colormaps 
top = cm.get_cmap('Oranges_r', 128) # r means reversed version
bottom = cm.get_cmap('Blues', 128)# combine it all
newcolors = np.vstack((top(np.linspace(0, 1, 128)),
                       bottom(np.linspace(0, 1, 128))))# create a new colormaps with a name of OrangeBlue
orange_blue = ListedColormap(newcolors, name='OrangeBlue')

plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = orange_blue)
plt.colorbar()
plt.show()
```

```python
# create colormap from 2 colors, not sequence 
N = 256

yellow = np.ones((N, 4))
yellow[:, 0] = np.linspace(255/256, 1, N) # R = 255
yellow[:, 1] = np.linspace(232/256, 1, N) # G = 232
yellow[:, 2] = np.linspace(11/256, 1, N)  # B = 11
yellow[:, 3] = np.linspace(0.9, 0.9, N)  # B = 11
yellow_cmp = ListedColormap(yellow)

print(yellow, yellow_cmp)

plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = yellow_cmp)
plt.colorbar()
plt.show()


red = np.ones((N, 4))
red[:, 0] = np.linspace(255/256, 1, N)
red[:, 1] = np.linspace(0/256, 1, N)
red[:, 2] = np.linspace(65/256, 1, N)
red_cmp = ListedColormap(red)

plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = red_cmp)
plt.colorbar()
plt.show()
```

```python
newcolors2 = np.vstack((yellow_cmp(np.linspace(0, 1, 128)),
                       red_cmp(np.linspace(1, 0, 128))))
double = ListedColormap(newcolors2, name='double')
plt.figure(figsize=(4, 3))
plt.pcolormesh(data, cmap=double)
plt.colorbar()
plt.show()
```

```python
# THIS IS THE BASIS TO CREATE!!!!!! Yeeeeiii
N = 256

my_color = np.ones((N, 4))
my_color[:, 0] = np.linspace(10/256, 10/256, N)  # R
my_color[:, 1] = np.linspace(10/256, 1, N)  # G
my_color[:, 2] = np.linspace(200/256, 1, N)   # B
my_color[:, 3] = np.linspace(0.9, 0.9, N)    # transparency
my_color_cmp = ListedColormap(my_color)

print(my_color, my_color_cmp)

plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = my_color_cmp)
plt.colorbar()
plt.show()
```

```python
# create colormap in subrange of existing one
# modified hsv in 256 color class
hsv_modified = cm.get_cmap('hsv', 256)# create new hsv colormaps in range of 0.3 (green) to 0.7 (blue)
newcmp = ListedColormap(hsv_modified(np.linspace(0.3, 0.7, 256)))# show figure
plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = newcmp)
plt.colorbar()
plt.show()

newcmp = ListedColormap(hsv_modified(np.linspace(0.0, 1.0, 256)))# show figure
plt.figure(figsize=(4,3))
plt.pcolormesh(data, cmap = newcmp)
plt.colorbar()
plt.show()
```
